### Plotting  from a script

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.show()
```

```python
x=np.linspace(0,10,100)

fig=plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

## Two interfaces for the price of one
A potentially confusing feature of Matplotlib is its dual interfaces: a convenient MATLAB-style state-based interface, and a more powerful object-oriented interface

```python
# MATLAB style interface

plt.figure() # create a plot figure

# create the first of two panels amd set current axis

plt.subplot(2,1,1) # (rows,cols,panel number )
plt.plot(x,np.sin(x))

# create the second panel and set current axis
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))

#It’s important to note that this interface is stateful:
# it keeps track of the “current” figure and axes,
#which are where all plt commands are applied.
```

## Object oriented interface
The object-oriented interface is available for these more complicated situations, and for when you want more control over your figure

```python
# First create a grid of plots
# ax will be an array of two axes objects
fig,ax=plt.subplots(2)

# call plot() method for appropriate object
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```
