```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# a simple sinusoid
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```python
# single figure with multiple lines
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

## Adjusting the plot : Line colors and styles
color can be specified in a variety of ways

```python
#controlling the color of plot elements
plt.plot(x,np.sin(x-0),color='blue')       #specify blue
plt.plot(x,np.sin(x-1),color='g')          #short code :rgbcmyk
plt.plot(x,np.sin(x-2),color='0.75')       #grayscale : 0-1
plt.plot(x,np.sin(x-3),color='#FFDD44')    #Hex code :RRGGBB :00-FF
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3))#RGB tuple :0-1
plt.plot(x,np.sin(x-5),color='chartreuse') #HTML color names
```

```python
# if no color specified ,automatically provide differnt color
plt.plot(x,x+0,linestyle='solid')
plt.plot(x,x+1,linestyle='dashed')
plt.plot(x,x+2,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted')

#for short use following code
plt.plot(x,x+4,linestyle='-')
plt.plot(x,x+5,linestyle='--')
plt.plot(x,x+6,linestyle='-.')
plt.plot(x,x+7,linestyle=':')

#example of various line style
```

```python
#linestyle and color codes can be combined into single monkeyword

plt.plot(x,x+0,'-g') #solid green
plt.plot(x,x+1,'--c') #dashed cyan
plt.plot(x,x+2,'-.k') #sllid black
plt.plot(x,x+3,':r') #dotted red

```

 single-character color codes reflect the standard abbreviations in the RGB (Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems, commonly used for digital color graphics

# Adjusting the plot : axes limits
Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes its nice to have finer control. The most basic way to adjust axis limits is to use the plt.xlim() and plt.ylim() method

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
# some cases ,need to reverse the axis ,reverse the order
plt.plot(x,np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

```python
'''
A useful related method is plt.axis() (note here the potential confusion between axes with an e,
and axis with an i). The plt.axis() method allows you to set the x and y limits with a single call, 
by passing a list that specifies [xmin, xmax, ymin, ymax] 

'''
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
# plt.axis() method goes even beyond
# automatically tighten bounds around the current plot

plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
# equal aspect ration so that on screen : x unit equal y unit
plt.plot(x,np.sin(x))
plt.axis('equal')
```

# Label plots

```python
plt.plot(x,np.sin(x))
plt.title('A sin curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
#multiple lines being shown within a single axis
#it useful to create a plot legend that each line type
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.axis('equal')
plt.legend()
```
