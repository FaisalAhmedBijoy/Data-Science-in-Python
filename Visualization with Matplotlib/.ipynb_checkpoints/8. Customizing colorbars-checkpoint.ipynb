```python
import numpy as np
import matplotlib.pyplot as plt 
```

Plot legends identify discrete labels of discrete points. For continuous labels based on the color of points, lines, or regions, a labeled colorbar can be a great tool. In Matplotlib, a colorbar is a separate axes that can provide a key for the meaning of colors in a plot.

```python
x=np.linspace(0,10,1000)
I=np.sin(x) * np.cos(x[:,np.newaxis])
plt.imshow(I)
plt.colorbar()
```

```python
# Customizing colorbars
plt.imshow(I,cmap='gray')
```

### choosing the colormap
1. sequential : one continious sequences of color :binary or viridis

2. Divergent : two distinct color ,positive and negative from mean (RdBu)

3. Qualitive : mix color with no particular sequences (rainbow or jet)

```python
from matplotlib.colors import LinearSegmentedColormap

def grayscale_cmap(cmap):
    cmap=plt.cm.get_cmap(cmap)
    colors=cmap(np.arange(cmap.N))
    
    # convert RGBA to perceived grayscale luminance
    RGB_weight=[0.299,0.587,0.114]
    luminance=np.sqrt(np.dot(colors[:,:3] **2 ,RGB_weight))
    colors[:,:3]=luminance[:,np.newaxis]
    
    return LinearSegmentedColormap.from_list(cmap.name+'_gray',colors,cmap.N)

def view_colormap(cmap):
    #plot a colormap with its grayscale equivalent
    
    cmap=plt.cm.get_cmap(cmap)
    colors=cmap(np.arange(cmap.N))
    
    cmap=grayscale_cmap(cmap)
    grayscale=cmap(np.arange(cmap.N))
    fig,ax=plt.subplots(2,figsize=(6,2),subplot_kw=dict(xticks=[],yticks=[]))
    ax[0].imshow([colors],extent=[0,10,0,1])
    ax[1].imshow([grayscale],extent=[0,10,0,1])

view_colormap('jet')
```

```python
view_colormap('viridis')
```

```python
view_colormap('cubehelix')
```

```python
view_colormap('RdBu')
```

## color limits and extensions
matplotlib allows for a large range of colorbar customization

```python
# an image that is subject to noise

I.shape
```

```python
#make noise is 1 % of the pixels
speckles=(np.random.random(I.shape)<0.01)
# print('Boolean dtype : speckles :',speckles[0][0])
#print(np.count_nonzero(speckles))

I[speckles]=np.random.normal(0,3,np.count_nonzero(speckles))
plt.figure(figsize=(10,3.5))

plt.subplot(1,2,1) # row,col,panel
plt.imshow(I,cmap='RdBu')
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(I,cmap='RdBu')
plt.colorbar(extend='both')
plt.clim(-1,1) # limit to indicate values
```

### Discreate colorbars
Colormaps are by default continuous, but sometimes you’d like to represent discrete values. The easiest way to do this is to use the plt.cm.get_cmap() function, and pass the name of a suitable colormap along with the number of desired bin

```python
print('A discreatized colormap')
#The discrete version of a colormap can be used just like any other colormap
plt.imshow(I,cmap=plt.cm.get_cmap('Blues',6))
plt.colorbar()
plt.clim(-1,1) # give the limit
```
