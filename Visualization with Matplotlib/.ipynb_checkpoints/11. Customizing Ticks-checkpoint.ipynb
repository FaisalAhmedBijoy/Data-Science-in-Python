## Major and Minor ticks
there is the concept of a major tick mark and a minor tick mark. As the names would imply, major ticks are usually bigger or more pronounced, while minor ticks are usually smaller. By default, Matplotlib rarely makes use of minor ticks, but one place you can see them is within logarithmic plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ax=plt.axes(xscale='log',yscale='log')
ax.grid()
```

```python
print("Logarithmic scales and labels")
print(ax.xaxis.get_major_locator())
print(ax.xaxis.get_minor_locator())

print(ax.xaxis.get_major_formatter())
print(ax.xaxis.get_minor_formatter())
```

We see that both major and minor tick labels have their locations specified by a LogLocator (which makes sense for a logarithmic plot). Minor ticks, though, have their labels formatted by a NullFormatter; this says that no labels will be shown

## Hiding ticks or labels
the most common tick/label formatting operation is the act of hiding ticks or labels. We can do this using plt.NullLocator() and plt.NullFormatter()

```python
ax=plt.axes()
ax.plot(np.random.rand(50))

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_locator(plt.NullLocator())
```

Notice that we’ve removed the labels (but kept the ticks/gridlines) from the x axis, and removed the ticks (and thus the labels as well) from the y axis. Having no ticks at all can be useful in many situations—for example, when you want to show a grid of images

## Hiding ticks within image plots in Handwritten digits


```python
fig,ax=plt.subplots(8,8,figsize=(6,6))
fig.subplots_adjust(hspace=0.2,wspace=0.2)
from sklearn.datasets import load_digits
digits=load_digits()

for i,axi in enumerate(ax.flat):
    axi.imshow(digits.images[i])
    axi.xaxis.set_major_locator(plt.NullLocator())
    axi.yaxis.set_major_locator(plt.NullLocator())
    axi.set()
```

Notice that each image has its own axes, and we’ve set the locators to null because the tick values (pixel number in this case) do not convey relevant information for this particular visualization.

## Reducing or Increasing the number of ticks

```python
fig,ax=plt.subplots(4,4,sharex=True,sharey=True)
```

Particularly for the x ticks, the numbers nearly overlap, making them quite difficult to decipher. We can fix this with the plt.MaxNLocator(), which allows us to specify the maximum number of ticks that will be displayed. Given this maximum number, Matplotlib will use internal logic to choose the particular tick location

```python
print("Customizing the number of ticks")
print("For every axis,set x and y major locator")

for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(3))
    axi.yaxis.set_major_locator(plt.MaxNLocator(3))
fig
```

## Fancy Tick Formats

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot a sin and cos curve
fig,ax=plt.subplots()
x=np.linspace(0,3*np.pi,1000)
ax.plot(x,np.sin(x),label='sin',lw=3)
ax.plot(x,np.cos(x),label='cos',lw=3)

#set up grid ,legend ,limits
ax.grid()
ax.legend()
ax.axis('equal')
ax.set_xlim(0,3 * np.pi)
```

There are a couple changes we might like to make. First, it’s more natural for this data to space the ticks and grid lines in multiples of π. We can do this by setting a Multi pleLocator, which locates ticks at a multiple of the number you provide. For good measure, we’ll add both major and minor ticks in multiples of π/4

```python
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))
#ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/4))

fig
```

```python
print("Multiplication of pi in  x axis")
def format_func(value,tick_number):
    print('value : ',value)
    # find number pf mutiple of pi /2
    N=int(np.round(2 * value /np.pi))
    print("value of N : ",N)
    if N==0 :
        return "0"
    elif N==1:
        return r"$\pi/2$"
    elif N==2:
        return r"$\pi$"
    elif N %2 >0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N//2)
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
        
    
```
